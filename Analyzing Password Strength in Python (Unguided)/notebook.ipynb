```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import re 
import numpy as np
```

```python
data = pd.read_csv("datasets/logins.csv",index_col = "id")
data.head()
```

```python
data['first_name'] = data.username.apply(lambda x: x.split('.')[0])
data['last_name'] = data.username.apply(lambda x: x.split('.')[1])
data.head()
```

```python
def pass_valid(password,first_name,last_name):
    lower_c = r'[a-z]+'
    upper_c = r'[A-Z]+'
    numeric_c = r'[0-9]+'
    non_numeric_c = r'\W+'
    pass_word = r'password'
    first_last_c = r'{0}|{1}'.format(first_name,last_name)
    
    regex1 = [lower_c,upper_c,numeric_c,non_numeric_c]
    regex2 = [pass_word,first_last_c]
    true_false1 = [bool(re.search(regex,password)) for regex in regex1]
    true_false2 = [bool(re.search(regex,password.lower())) for regex in regex2]
    
    if (False in true_false1) or (True in true_false2) or (len(password)<10):
        return False
    else:
        return True
    
    
    
    
```

```python
bad_pass_counter = 0
email_list = []
for password,first,last,username in zip(data.password,data.first_name,data.last_name,data.username):
    if pass_valid(password,first,last) == False:
        bad_pass_counter+=1
        email_list.append(username)
        
bad_pass = round(bad_pass_counter/len(data),2)
email_list = pd.Series(email_list).sort_values()
```
